```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy
import time
from os import path
```

```python
url = 'https://www.kijiji.ca/b-for-rent/kingston-on'
#url = 'https://www.kijiji.ca/b-for-rent/kingston' # This works too!
baseurl = 'https://www.kijiji.ca'
baseForKingston = '/c30349001l1700183'

pageNos = '/page-'
# A sample URL would look like 
# https://www.kijiji.ca/b-for-rent/ontario/page-8/c30349001l9004

# Using
# https://www.kijiji.ca/b-for-rent/kingston-on/c30349001l1700183
```

```python
numPages = 10
adurl = []
linksFromText = ""
```

```python
def saveLinks(adurl):
    f = open("links.txt", "w")
    out = ""
    
    for ad in adurl:
        out += ad + "\n"
        
    f.write(out)
        
    f.close()
        
def getDetails(adurl):
    f = open('links.txt')
    
    for line in f:
        print(line[0:-1])
        
    f.close()
```

```python
# I saved those links into a textFile to avoid duplicated / to 
# reduce number of hits on the server

# if file exists read the Advt URLs from text file and 
# perform the scraping

if path.exists('links.txt'):
    with open('links.txt', 'r') as f:
        linksFromText = f.readlines()
    #print(linksFromText)
    f.close()
    # A fucntion call to scrape the actual data
    getDetails(linksFromText)
# if file does not exists perform the scraping
else:
    for i in range(numPages):
        
        url_final = url+pageNos+str(i)+baseForKingston
        response = requests.get(url_final)
        soup = BeautifulSoup(response.text, "html.parser")
        advtTitles = soup.findAll('div', attrs={'class' : 'title'})
        try:
            for link in advtTitles:
                adlink = baseurl+link.find('a')['href']
                adurl.append(adlink)
        except Exception as e:
              print(e)
        time.sleep(1)
        
    # a fuction to write the 
    saveLinks(adurl)
    
    # a fuction to do the actual scraping
    getDetails(adurl)
```

```python
# Should be setup like this
# adInfo = [{url: ---, title: ---, price: ---, description: ---, location: ---, date : ---},
#           {url: ---, title: ---, price: ---, description: ---, location: ---, date : ---},
#           {url: ---, title: ---, price: ---, description: ---, location: ---, date : ---},
#           ...
#           {url: ---, title: ---, price: ---, description: ---, location: ---, date : ---}]
adInfos = []

for url in linksFromText:
    listDetails = ""
    url = url.rstrip('\n')
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    
    adDic = {}
    
    try:
        adDic['url'] = url
        adDic['title'] = soup.select_one("h1[class*=title-2323565163]").text
        pr = soup.select_one("span[class*=currentPrice-2842943473]").text
        adDic['price'] = pr[1:pr.index(".")]
        #adDic['desc'] = soup.find_all('div', attrs={'class' : 'descriptionContainer-3544745383'})
        adDic['loc'] = soup.find('span', attrs={'class' : 'address-3617944557'}).text
        adDic['time'] = soup.find('time')['datetime'][:10].replace("-", "")
        
        adInfos.append(adDic)
    except:
        # Wasn't successful in getting info
        pass
    
    # Doesn't do it right away, please don't shadowban me
    time.sleep(.01)
```
