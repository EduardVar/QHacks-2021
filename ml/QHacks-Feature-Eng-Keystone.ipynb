```python
# Custom cols
cols =list(data[0].keys())
cols = cols[:-1] + ["A/C", "electric", "heat"]

# Build the dataframe
clean_df = pd.DataFrame(columns=cols)
                
i = 0
for line in data:
    
    price = 0
    try:
        price = round(Decimal(sub(r'[^\d.]', '', line["price"][1:line["price"].index("/")])))
    except:
        price = np.nan
      
    date = ""
    split_date = line['date-available'].split()
    try:
        date = str(random.randint(1,2)) + str(random.randint(1,9)) + monthToNum[split_date[0]] + split_date[1]
        print(date)
    except:
        date = datetime.today().strftime('%d%m%Y')
    
    clean_df.loc[i] = [line['address'], line['bedrooms'], date, 
                       line['landlords'], price, line['title'], line['url'], 
                       int(line["utilites"]["A/C"]), 
                       int(line["utilites"]["electric"]),
                       int(line["utilites"]["heat"])]
    
    i += 1
```
