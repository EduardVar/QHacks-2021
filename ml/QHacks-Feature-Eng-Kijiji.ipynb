```python
import pandas as pd 
import numpy as np

from re import sub
from decimal import Decimal
from datetime import datetime

import json 
import csv 

# Opening JSON file and loading the data 
# into the variable data 
with open('data/kijiji-output.json') as json_file: 
    data = json.load(json_file) 
```

```python
# Keywords
bedId = ["bed", "bedroom", "bedrooms","bdrm", "beds"]
bathId = ["bath", "baths", "bathroom", "bathrooms", "rest", "rests", "restroom", "restrooms"]
minuteId = ["min", "mins", "minutes", "minute", "minutes"]

bad = ["food", "promo", "commercial"]

textDic = {"one": 1, "two": 2, "three": 3, "four": 4, "five":5, 
           "six": 6, "seven": 7, "eight": 8, "nine": 9, "ten": 10}
```

```python
bds = 0
smollTitle = data[0]['title'].lower().replace("-", " ")
smollTitle = smollTitle.split()
for j in range(len(smollTitle)):
    if smollTitle[j] in bedId:  

        prev = smollTitle[j-1]

        try:
            bds = int(prev)
        except:    
            if prev in textDic:
                bds = textDic[prev]
            else:
                print("FAILED Distance calc")
                bds = np.nan

        break
            
print(bds)
```

```python
# Build the dataframe
clean_df = pd.DataFrame(columns=list(data[0].keys())+["beds","baths"])
bad_words = ["food"]

i = 0

for line in data:
    # Checks if it's a non rental posting
    if any(sub in data[i]['title'].lower() for sub in bad):
        print("WATCH OUT")
        i += 1
        continue
    
    # The dirty string
    dirty = list(line.values())
    
    # For money
    money = 0
    try:
        money = round(Decimal(sub(r'[^\d.]', '', dirty[0])))
    except:
        money = np.nan
        
    # For title
    title = dirty[1]
        
    # For distance
    distance = 20
    lTitle = title.lower().replace("-", " ")
    ltSplit = lTitle.split()
    
    for j in range(len(ltSplit)):
        if ltSplit[j] in minuteId:
            if ltSplit[j] == "minutes":
                distance = 3
                break
            
            prev = ltSplit[j-1]

            try:
                distance = int(prev)
            except:    
                if prev in textDic:
                    distance = textDic[prev]
                else:
                    #print("FAILED Distance calc")
                    distance = 20
                    
            break
        
    # Find beds
    beds = 1
    for j in range(len(ltSplit)):
        if ltSplit[j] in bedId:  
            prev = ltSplit[j-1]

            try:
                beds = int(prev)
            except:    
                if prev in textDic:
                    beds = textDic[prev]
                else:
                    #print("FAILED Bed calc")
                    beds = np.nan

            break
            
    # Find bathrooms
    baths = 1
    for j in range(len(ltSplit)):
        if ltSplit[j] in bathId:  
            prev = ltSplit[j-1]

            try:
                baths = int(prev)
            except:    
                if prev in textDic:
                    baths = textDic[prev]
                else:
                    #print("FAILED Bed calc")
                    baths = np.nan

            break
        
    # For date
    date = dirty[3].replace("/", "")
    
    # Makes it today's date if it got posted recently
    if len(date) == 0 or date[0] == "<":
        date = datetime.today().strftime('%d%m%Y')

    #print(lTitle)
    
    clean_df.loc[i] = [money, data[i]['title'], distance,
                       date, data[i]['location'], data[i]['description'],
                       beds, baths]
    
    i += 1
```
